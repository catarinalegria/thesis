```python
from sklearn.feature_extraction.text import TfidfVectorizer
import re
import pandas as pd
from nltk.tokenize import sent_tokenize
import csv
import string

import pandas as pd
import numpy as np
```

```python
#docs = process_file_friends("1x01 - The One Where Monica Gets A Roommate.srt")
file = open("friends.csv", 'r', encoding="utf8")
csvreader = csv.reader(file)
data = list(csvreader)
characters =["Rachel", "Monica", "Joey", "Chandler", "Ross", "Phoebe"]
    
vectorizer = TfidfVectorizer()
```

```python
values = []
values_lower = []

for character in characters:
    lines = []
    #create auxiliar vector WITHOUT punctuation for TF-IDF purposes
    sp_character = []


    for row in data:
        if row[-2] == character:
        #if row[-2] == character and row[0]=="S01" and row[1]=="E01":
            #we only want dialogue
            lines.append(row[-1])

        
    for line in lines:
        #remve punctuation
        line = line.translate(str.maketrans('', '', string.punctuation))
        sp_character.append(line.lower())

    sp_character = list( dict.fromkeys(sp_character) )

    '''with open('aux1.txt', 'w', encoding="utf-8") as f:
        for elem in sp_character:
            f.write(elem + "\n")'''

    #compute tf-idf
    tf_idf = vectorizer.fit_transform(sp_character)
    tf_idf_array = tf_idf.toarray()
    scores = []
        
    for sentence in tf_idf_array:
        value = 0
        for n in sentence:
            value+=n
        scores.append(value)

    final = []

    #format: final = [...,[sentence, tf-idf score],...]
    for i in range(len(scores)):
        cell = []
        #uses original array with punctuation
        cell= [scores[i], sp_character[i]]
        final.append(cell)
    
    # lower -> higher     
    sorted_list = sorted(final)
    l = []

    #get only the strings
    for elem in sorted_list:
        l.append(elem[1])

    #remove duplicates
    l = list( dict.fromkeys(l) )
    
    counts = l[-15:]
    lower = l[:20]

    values.append(counts)
    values_lower.append(lower)

dict = {key: value for key, value in zip(characters, values)}
```
